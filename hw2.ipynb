## 问题 1  

> 给定一个已排序的链表的头 `head` ， *删除所有重复的元素，使每个元素只出现一次* 。返回 *已排序的链表* 。链表的类如下所示：

```python
class NodeList:
    def __init__(self, val=None, right=None):
        self.val   = val
        self.right = right
```

输入是一个数组，你首先需要将数组转化为链表，然后删除链表中的重复元素，再遍历链表元素，以一个数组的形式返回。请设计一个算法解决上述任务，分析算法设计思路，计算时间复杂度, 并基于python编程实现。

e.g.  输入：head=[1, 1, 2, 3, 3]   输出：[1, 2, 3]

![image-20240502110020439](./fig/hw2q1.png)



**idea：** 删除链表：遍历链表时判断下一个节点值和下下一个节点值是否相同，如果相同，直接跳过下一个节点

````python
# add your idea here
class NodeList:
    def __init__(self, val=None, right=None):
        self.val = val
        self.right = right

def check(arr):
    #数组转换为链表
    head = NodeList(arr[0])
    current = head
    for val in arr:
        current.right = NodeList(val)
        current = current.right
    #删除多余元素
    current = head
    while current and current.right:
        if current.val == current.right.val:
            current.right = current.right.right
        else:
            current = current.right
    #重新转化为数组
    result = []
    current = head
    while current:
        result.append(current.val)
        current = current.right
    return result

print(check([1, 1, 2, 3, 3]))#输出[1, 2, 3]
# your algorithm time complexity is:O(n)
````

输入：nums=[2,7,11,15], target=9， 输出：[0，1]

输入：nums=[3,2,4], target=6, 输出：[1,2]

**idea：** 生成一个字典，遍历每个元素，如果元素对应的目标元素在字典中存在，则返回，否则加入到字典中

````python
# add your code here
def two_sum(nums, target):
    num_map = {}  
    for index, num in enumerate(nums):
        temp = target - num  
        if temp in num_map:  
            return [num_map[temp], index]  
        num_map[num] = index  
    return []  
print(check([2,7,11,15]))# 输出：[0, 1]
# your algorithm time complexity is: O(n)
````

>
> 以表达式 $3+5 \times 8-6$​ 为例。编译器使用两个栈来完成运算，即一个栈保持操作数，另一个栈保存运算符。

**idea：** 使用栈存储元素和操作符进行运算

````python
# add your code here 借助DeepSeek模型
def check(str):
    def operation(op,a,b):
        if op == '+':
            return a+b
        elif op == '-':
            return a-b
        elif op == '*':
            return a*b
        elif op == '/':
            return a//b  
        else:
            return 0

    def Priority(op):
        if op=='+' or op=='-':
            return 1
        elif op=='*' or op=='/':
            return 2
        return 0

    operand_stack=[]
    operator_stack=[]

    i = 0
    while i < len(str):
        if str[i].isdigit():
            num=0
            while i<len(str)and str[i].isdigit():
                num=num*10+int(str[i])
                i += 1
            operand_stack.append(num)
            i -= 1  
        elif str[i] in '+-*/':
            while (operator_stack and
                   Priority(operator_stack[-1]) >= Priority(str[i])):
                op = operator_stack.pop()
                b = operand_stack.pop()
                a = operand_stack.pop()
                operand_stack.append(Priority(op, a, b))
            operator_stack.append(str[i])
        i += 1

    while operator_stack:
        op=operator_stack.pop()
        b=operand_stack.pop()
        a=operand_stack.pop()
        operand_stack.append(operation(op, a, b))

    return operand_stack[0]
print(check("3+5*8-6"))# 输出：37
# your algorithm time complexity is: O(n)
````

>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]



**idea：** 用栈存储星球元素，遇到正数入栈，正数后遇到负数就计算结果

````python
def check(nums):
    if not nums:  # 检查输入是否为空
        return []
    stack = [0]
    for num in nums:
        while stack and num < stack[-1]:
            stack.pop()
        if stack and stack[-1] == num:
            stack.pop()
        else:
            stack.append(num)
    return stack
print(check([-3,-6,2,8, 5,-8,9,-2,1]))#输出：[-3, -6, 2, 8, 9, 1]
#时间复杂度:O(n)
````

> 给定由n个正整数组成的一个集合$S = \{a_1, a_2，···，a_n\}$和一个正整数W，设计一个算法确定是否存在S的一个子集 $K \subseteq S$, 使K中所有数之和为 $W$, 如果存在返回“True”，否则返回“False”

请设计一个时间复杂度为$O(nW)$动态规划算法，解决上述问题，分析算法的设计思路，并且基于python编程实现（不需要输出子集）。

e.g. 

输入：S = {1,4,7,3,5}， W = 11，输出：True。   因为K可以是{4,7}。



**idea：** dp[i][j] 表示前 i 个元素中是否存在一个子集的和为 j，true表示存在，false表示不存在

````python
# add your code here 借助DeepSeek
def check(s, w):
    n = len(s)
    dp = [[False]*(w+1) for _ in range(n+1)]
    for i in range(n+1):
        dp[i][0] = True
    for i in range(1,n+1):
        for j in range(1,w+1):
            if j < s[i-1]:
                dp[i][j]=dp[i-1][j]
            else:
                dp[i][j]=dp[i-1][j] or dp[i-1][j-s[i-1]]
    return dp[n][w]
print(check([1,4,7,3,5],11))#输出：true
# your algorithm time complexity is: O(nW)
````

> 给定一个n个物品的集合。物体的重量为$w_1, w_2，…、w_n$，物品的价值分别是$v_1、v_2、…v_n$。给你**两个**重量为 $c$ 的背包。如果你带了一个东西，它可以放在一个背包里，也可以放在另一个背包里，但不能同时放在两个背包里。所有权重和价值都是正整数。

1. 设计一个时间复杂度为 $O(nc^2)$ 的动态规划算法，确定可以放入两个背包的物体的最大价值。分析算法设计思路，并基于python编程实现
2. \* 修改1中的算法，输出每个背包的内容（物品对应下标）。

e.g.: 

输入 V=[1,3,2,5,8,7], W=[1,3,2,5,8,7], c=7, 输出：最大价值=14，背包装的物品为：[6] [4，3] （同一个背包中物品装入顺序对结果无影响）  


**idea：** pre[i][j]和current[i][j]分别表示当前和之前在背包1使用 w1 容量、背包2使用 w2 容量时的最大价值。通过遍历每个物品，选择放入背包1、背包2或不放，更新状态背包1使用w1容量和背包2使用w2

````python
# add your idea here 借助DeepSeek模型
def check(w, v, c):
    n = len(w)
    prev = [[-float('inf')] * (c + 1) for _ in range(c + 1)]
    prev[0][0] = 0
    for i in range(n):
        current = [[-float('inf')] * (c + 1) for _ in range(c + 1)]
        w_i, v_i = w[i], v[i]
        for w1 in range(c + 1):
            for w2 in range(c + 1):
                if prev[w1][w2]==-float('inf'):
                    continue
                if current[w1][w2]<prev[w1][w2]:
                    current[w1][w2]=prev[w1][w2]
                new_w1 = w1 + w_i
                if new_w1 <= c and current[new_w1][w2] < prev[w1][w2] + v_i:
                    current[new_w1][w2] = prev[w1][w2] + v_i
                new_w2 = w2 + w_i
                if new_w2 <= c and current[w1][new_w2] < prev[w1][w2] + v_i:
                    current[w1][new_w2] = prev[w1][w2] + v_i
        prev = current

    max_val = max(prev[w1][w2] for w1 in range(c + 1) for w2 in range(c + 1))
    return max_val if max_val != -float('inf') else 0
print(check([1,3,2,5,8,7],[1,3,2,5,8,7],7))#输出14
# your algorithm time complexity is: O(nc^2)
````

改进算法：在动态规划过程中保存历史状态，回溯时根据最大价值逆向推导物品选择路径 (借助DeepSeek模型)

````python
def check(w, v, c):
    n = len(w)
    history = []
    prev = [[-float('inf')] * (c + 1) for _ in range(c + 1)]
    prev[0][0] = 0
    history.append([row.copy() for row in prev])

    for i in range(n):
        current = [[-float('inf')] * (c + 1) for _ in range(c + 1)]
        w_i, v_i = w[i], v[i]
        for w1 in range(c + 1):
            for w2 in range(c + 1):
                if prev[w1][w2] == -float('inf'):
                    continue
                if current[w1][w2] < prev[w1][w2]:
                    current[w1][w2] = prev[w1][w2]
                new_w1 = w1 + w_i
                if new_w1 <= c and current[new_w1][w2] < prev[w1][w2] + v_i:
                    current[new_w1][w2] = prev[w1][w2] + v_i
                new_w2 = w2 + w_i
                if new_w2 <= c and current[w1][new_w2] < prev[w1][w2] + v_i:
                    current[w1][new_w2] = prev[w1][w2] + v_i
        prev = [row.copy() for row in current]
        history.append([row.copy() for row in prev])

    # 找最大价值和对应的背包容量
    max_val = -float('inf')
    best_w1, best_w2 = 0, 0
    for w1 in range(c + 1):
        for w2 in range(c + 1):
            if prev[w1][w2] > max_val:
                max_val = prev[w1][w2]
                best_w1, best_w2 = w1, w2

    if max_val == -float('inf'):
        return 0, [], []

    # 回溯路径
    bag1, bag2 = [], []
    current_w1, current_w2 = best_w1, best_w2
    for i in range(n - 1, -1, -1):
        current_val = history[i + 1][current_w1][current_w2]
        prev_vals = history[i]
        # 检查是否不选
        if prev_vals[current_w1][current_w2] == current_val:
            continue
        # 检查是否放入背包1
        if current_w1 >= w[i] and prev_vals[current_w1 - w[i]][current_w2] + v[i] == current_val:
            bag1.append(i)
            current_w1 -= w[i]
        # 检查是否放入背包2
        elif current_w2 >= w[i] and prev_vals[current_w1][current_w2 - w[i]] + v[i] == current_val:
            bag2.append(i)
            current_w2 -= w[i]

    bag1.reverse()
    bag2.reverse()
    return max_val, bag1, bag2

````

## 问题 10 

> 给定两个字符串 $x[1..n]$ 和 $y[1..m]$，我们想通过以下操作将 $x$ 变换为 $y$ :
>
> **插入**：在 $x$ 中插入一个字符(在任何位置)；**删除**：从 $x$ 中删除一个字符(在任何位置)； **替换**：用另一个字符替换 $x$ 中的一个字符。
>
> 例如: $x = abcd$, $y = bcfe$，
>
> - 将 $x$ 转换为 $y$ 的一种可能方法是：1. 删除 $x$ 开头的 $a$, $x$变成 $bcd$； 2. 将 $x$ 中的字符 $d$ 替换为字符 $f$。$x$ 变成 $bcf$； 3. 在 $x$ 的末尾插入字符 $e$。$x$ 变成 $bcfe$。
>
> - 另一种可能的方法：1. 删除 $x$ 开头的 $a$,  $x$ 变成 $bcd$； 2. 在 $x$ 中字符 $d$ 之前插入字符 $f$。$x$ 变成 $bcfd$。3. 将 $x$ 中的字符 $d$ 替换为字符 $e$。$x$ 变成 $bcfe$。

设计一个时间复杂度为 $O(mn)$ 的算法，返回将 $x$ 转换为 $y$ 所需的最少操作次数。分析算法设计思路，并基于python编程实现。


**idea：** 使用一个二维数组 dp，其中 dp[i][j] 表示将字符串 x 的前 i 个字符转换为字符串 y 的前 j 个字符所需的最少操作次数，
如果 x[i-1] == y[j-1]，dp[i][j] = dp[i-1][j-1]（无需操作）。
否则：
替换：dp[i][j] = dp[i-1][j-1] + 1
删除：dp[i][j] = dp[i-1][j] + 1
插入：dp[i][j] = dp[i][j-1] + 1
dp[i][j] 取这三种操作中的最小值。

````python
# add your idea here 借助DeepSeek模型
def check(x, y):
    n=len(x)
    m=len(y)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(n+1):
        dp[i][0] = i
    for j in range(m+1):
        dp[0][j] = j
    for i in range(1, n+1):
        for j in range(1, m+1):
            if x[i-1] == y[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                replace = dp[i-1][j-1]+1
                delete = dp[i-1][j]+1
                insert = dp[i][j-1]+1
                dp[i][j] = min(replace, delete, insert)
    return dp[n][m]

# your algorithm time complexity is: O(mn)
````
